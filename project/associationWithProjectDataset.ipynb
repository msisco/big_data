```python
# Take the array and make it and transform all of its elements into identifiers to be used in the
# Apriori Association algorithm.
```

```python
def makeBaskets(array):
    for x in range(0, len(array)):
        # 0 = "low_duration"
        # 0-48 = "medium_duration"
        # >48 = "high_duration"
        if array[x][0]== 0:
            array[x][0] = "low_duration"
        elif array[x][0] > 0 and array[x][0]<= 48:
            array[x][0] = "medium_duration"
        elif array[x][0] > 48:
            array[x][0] = "high_duration"
            
        # 0-520 = low_src_bytes
        # 520 - 3025 = medium_src_bytes
        # >3025 = high_src_bytes
        if array[x][4] <=520:
            array[x][4] = "low_src_bytes"
        elif array[x][4] > 520 and array[x][4] <= 3025:
            array[x][4] = "medium_src_bytes"
        elif array[x][4] > 3025:
            array[x][4] = "high_src_bytes"
            
        # 0 = low_dst_bytes
        # 0-869 = medium_dst_bytes
        # > 869 = high_dst_bytes
        if array[x][5] == 0:
            array[x][5] = "low_dst_bytes"
        elif array[x][5]> 0 & array[x][5]<= 869:
            array[x][5] = "medium_dst_bytes"
        elif array[x][5]>869:
            array[x][5] = "high_dst_bytes"
            
        # yes_land or no_land
        if array[x][6] == 0:
            array[x][6] = "no_land"
        elif array[x][6] == 1:
            array[x][6] = "yes_land"
            
        # 0 = low_wrong_fragment
        # 0-1 = medium_wrong_fragment
        # >1 = high_wrong_fragment
        if array[x][7] == 0:
            array[x][7] = "low_wrong_fragment"
        elif array[x][7] >0 & array[x][7] <= 1:
            array[x][7] = "medium_wrong_fragment"
        elif array[x][7] > 1:
            array[x][7] = "high_wrong_fragment"
            
        # 0 = low_urgent
        #>0 = high_urgent
        if array[x][8] == 0:
            array[x][8] = "low_urgent"
        elif array[x][8] > 0:
            array[x][8] = "high_urgent"
            
        # 0 = low_hot
        # 0-1 = medium_hot
        # >0 = high_hot
        if array[x][9] == 0:
            array[x][9] = "low_hot"
        elif array[x][9] >0 & array[x][9] <=1:
            array[x][9] = "medium_hot"
        elif array[x][9] >1:
            array[x][9] = "high_hot"
            
        # 0 = low_num_failed_logins
        # 0-1 = medium_num_failed_logins
        # >1 = high_num_failed_logins
        if array[x][10] == 0:
            array[x][10] = "low_num_failed_logins"
        elif array[x][10] >0 & array[x][10]<=1:
            array[x][10] = "medium_num_failed_logins"
        elif array[x][10] >1:
            array[x][10] = "high_num_failed_logins"
            
        #no_logged_in
        #yes_logged_in
        if array[x][11] == 0:
            array[x][11] == "no_logged_in"
        elif array[x][11] == 1:
            array[x][11] = "yes_logged_in"
            
        # 0 = low_num_compromised
        # >0 = high_num_compromised
        if array[x][12] == 0:
            array[x][12] = "low_num_compromised"
        elif array[x][12] >0:
            array[x][12] = "high_num_compromised"
            
        # 0 = low_root_shell
        # >0 = high_root_shell
        if array[x][13] == 0:
            array[x][13] = "low_root_shell"
        elif array[x][13] >0:
            array[x][13] = "high_root_shell"
            
        #0 = low_su_attempted
        #>1 = high_su_attempted
        if array[x][14] == 0:
            array[x][14] = "low_su_attempted"
        elif array[x][14] >0:
            array[x][14] = "high_su_attempted"
            
        #0 = low_num_root
        # >0 = high_num_root
        if array[x][15] == 0:
            array[x][15] = "low_num_root"
        elif array[x][15] >0:
            array[x][15] = "high_num_root"
            
        # 0 = low_file_creations
        #>1 = high_file_creations
        if array[x][16] == 0:
            array[x][16] = "low_file_creations"
        elif array[x][16] >0:
            array[x][16] = "high_file_creations"
            
        # 0 = low_num_shells
        # >1 = high_num_shells
        if array[x][17] == 0:
            array[x][17] = "low_num_shells"
        elif array[x][17] >0:
            array[x][17] = "high_num_shells"
            
        # 0 = low_num_access_files
        #>1 = high_num_access_files
        if array[x][18] == 0:
            array[x][18] = "low_num_access_files"
        elif array[x][18] >0:
            array[x][18] = "high_num_access_files"
            
        # 0 = low_num_outbound_cmds
        #>1 = high_num_outbound_cmds
        if array[x][19] == 0:
            array[x][19] = "low_num_outbound_cmds"
        elif array[x][19] >0:
            array[x][19] = "high_num_outbound_cmds"
             
        #'is_host_login',
        if array[x][20] == 0:
            array[x][20] = "no_host_login"
        elif array[x][20] ==1:
            array[x][20] = "yes_host_login"
 
        #'is_guest_login',
        if array[x][21] == 0:
            array[x][21] = "no_guest_login"
        elif array[x][21] ==1:
            array[x][21] = "yes_guest_login"
            
        #0-332 low_count
        #332-510 medium_count
        #510-511 high_count
        if array[x][22] < 332:
            array[x][22] = "low_count"
        elif array[x][22]>= 332 and array[x][22]<510:
            array[x][22] = "medium_count"
        elif array[x][22]>= 510:
            array[x][22] = "high_count"
            
        #0-293 = low_srv_count
        #293-510 = medium_srv_count
        #>=510 = high_srv_count
        if array[x][23] < 292:
            array[x][23] = "low_srv_count"
        elif array[x][23]>= 292 and array[x][23] <510:
            array[x][23] = "medium_srv_count"
        elif array[x][23]>= 510:
            array[x][23] = "high_srv_count"
            
        # 0 = low_serror_rate
        # 0 - 0.17 = medium_serror_rate
        # > 0.17 = high_serror_rate
        
        if array[x][24] <0.01:
            array[x][24] = "low_serror_rate"
        elif array[x][24] >=0.01 and array[x][24] <= 0.17:
            array[x][24] = "medium_serror_rate"
        elif array[x][24] > 0.17:
            array[x][24] = "high_serror_rate"
            
        # 0 = low_srv_serror_rate
        # 0 - 0.17 = medium_srv_serror_rate
        # > 0.17 = high_srv_serror_rate
        if array[x][25] <0.01:
            array[x][25] = "low_srv_serror_rate"
        elif array[x][25] >=0.01 and array[x][25] <= 0.17:
            array[x][25] = "medium_srv_serror_rate"
        elif array[x][25] > 0.17:
            array[x][25] = "high_srv_serror_rate"
            
        # 0 = low rerror_rate
        # 0 - 0.05 = medium_rerror_rate
        #>.05 = high_rerror_rate
        if array[x][26] <0.01:
            array[x][26] = "low_rerror_rate"
        elif array[x][26] >=0.01 and array[x][26] <= 0.05:
            array[x][26] = "medium_rerror_rate"
        elif array[x][26] > 0.05:
            array[x][26] = "high_rerror_rate"
            
        # 0 = low_srv_rerror_rate
        # 0 - 0.05 = medium_srv_rerror_rate
        #>.05 = high_srv_rerror_rate
        if array[x][27] <0.01:
            array[x][27] = "low_srv_rerror_rate"
        elif array[x][27] >=0.01 and array[x][27] <= 0.05:
            array[x][27] = "medium_srv_rerror_rate"
        elif array[x][27] > 0.05:
            array[x][27] = "high_srv_rerror_rate"
            
        # 0- .79 = low_same_srv_rate
        #.79-1 = medium_same_srv_rate
        # 1 = high_same_srv_rate
        if array[x][28] <.79:
            array[x][28] = "low_same_srv_rate"
        elif array[x][28] >=0.79 and array[x][28] <1:
            array[x][28] = "medium_same_srv_rate"
        elif array[x][28] >= 1:
            array[x][28] = "high_same_srv_rate"
            
        # 0 = low_diff_srv_rate
        # 0-.02 = medium_diff_srv_rate
        # > .02 = high_diff_srv_rate
        if array[x][29] <.001:
            array[x][29] = "low_diff_srv_rate"
        elif array[x][29] >=.001 and array[x][29] <.02:
            array[x][29] = "medium_diff_srv_rate"
        elif array[x][29] >= .02:
            array[x][29] = "high_diff_srv_rate"
            
        # 0 = low_srv_diff_host_rate
        # 0-.02 = medium_srv_diff_host_rate
        # > .02 = high_srv_diff_host_rate
        if array[x][30] <.001:
            array[x][30] = "low_srv_diff_host_rate"
        elif array[x][30] >=.001 and array[x][30] <.02:
            array[x][30] = "medium_srv_diff_host_rate"
        elif array[x][30] >= .02:
            array[x][30] = "high_srv_diff_host_rate"
            
        #255 = high_dst_host_count
        #255-232 = medium_dst_host_count
        # 0-232 = low_dst_host_count
        if array[x][31] <232:
            array[x][31] = "low_dst_host_count"
        elif array[x][31] >=232 and array[x][31] <=254:
            array[x][31] = "medium_dst_host_count"
        elif array[x][31] > 254:
            array[x][31] = "high_dst_host_count"
            
        #255 = high_dst_host_srv_count
        #255-188 = medium_dst_host_srv_count
        # 0-188 = low_dst_host_srv_count
        if array[x][32] <188:
            array[x][32] = "low_dst_host_srv_count"
        elif array[x][32] >=188 and array[x][32] <=254:
            array[x][32] = "medium_dst_host_srv_count"
        elif array[x][32] > 254:
            array[x][32] = "high_dst_host_srv_count"
        
        # 0-.75 low_dst_host_same_srv_rate
        # .75 -1 high_dst_host_sam_srv_rate
        
        if array[x][33]<.75:
            array[x][33] = "low_dst_host_same_srv_rate"
        elif array[x][33] >= .75:
            array[x][33] = "high_dst_host_sam_srv_rate"
            
        # 0 low_dst_host_diff_srv_rate
        # >0 high_dst_host_diff_srv_rate
        if array[x][34]<.01:
            array[x][34] = "low_dst_host_diff_srv_rate"
        elif array[x][34] >=.01:
            array[x][34] = "high_dst_host_diff_srv_rate"
            
        #<.6 low_dst_host_same_src_port_rate
        #>=.6 high_dst_host_same_src_port_rate
        if array[x][35]<.6:
            array[x][35] = "low_dst_host_same_src_port_rate"
        elif array[x][35] >=.6:
            array[x][35] = "high_dst_host_same_src_port_rate"
        
        # <.01 low_dst_host_srv_diff_host_rate
        # >.01 high_dst_host_srv_diff_host_rate
        if array[x][36]<.01:
            array[x][36] = "low_dst_host_srv_diff_host_rate"
        elif array[x][36] >=.01:
            array[x][36] = "high_dst_host_srv_diff_host_rate"
            
        # <.17 low_dst_host_serror_rate
        # >.17 high_dst_host_serror_rate
        if array[x][37]<.17:
            array[x][37] = "low_dst_host_serror_rate"
        elif array[x][37] >=.17:
            array[x][37] = "high_dst_host_serror_rate"
            
        # <.17 low_dst_host_srv_serror_rate
        #>.17 high_dst_host_srv_serror_rate
        if array[x][38]<.17:
            array[x][38] = "low_dst_host_srv_serror_rate"
        elif array[x][38] >=.17:
            array[x][38] = "high_dst_host_srv_serror_rate"
        
        # <.05 low_dst_host_rerror_rate
        # >.05 high_dst_host_rerror_rate
        if array[x][39]<.05:
            array[x][39] = "low_dst_host_srv_serror_rate"
        elif array[x][39] >=.05:
            array[x][39] = "high_dst_host_srv_serror_rate"
            
        #<.05 low_dst_host_srv_rerror_rate
        #>.05 high_dst_host_srv_rerror_rate
        if array[x][40]<.05:
            array[x][40] = "low_dst_host_srv_serror_rate"
        elif array[x][40] >=.05:
            array[x][40] = "high_dst_host_srv_serror_rate"
        
```
